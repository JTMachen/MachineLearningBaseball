```python
data = pd.read_csv("Statcast.csv")
expected_bat = pd.read_csv("Expected_Batter.csv")
name_list = []
for x in range(0,len(expected_bat)):
    name_list.append(expected_bat['first_name'][x][1:] + " " + expected_bat['last_name'][x])
expected_bat['full_name'] = name_list
player_dict = dict(zip(expected_bat['player_id'],expected_bat['full_name']))
for batter in data['batter'].unique():
    try:
        player_dict[batter]
    except:
        data = data[data['batter'] != batter]
data['batter'] = data['batter'].apply(lambda x: player_dict[x])
## Changing pitcher column in data to pitcher name in easier to read format
expected_pitch = pd.read_csv("Expected_Pitcher.csv")
name_list = []
for x in range(0,len(expected_pitch)):
    name_list.append(expected_pitch['first_name'][x][1:] + " " + expected_pitch['last_name'][x])
expected_pitch['full_name'] = name_list
player_dict = dict(zip(expected_pitch['player_id'],expected_pitch['full_name']))
for batter in data['pitcher'].unique():
    try:
        player_dict[batter]
    except:
        data = data[data['pitcher'] != batter]
data['pitcher'] = data['pitcher'].apply(lambda x: player_dict[x])
data.head(1)
```

```python
data.events.unique()
```

```python
results_df = data.fillna(0)
total_bat_df = pd.DataFrame()
results_df = results_df[results_df['events'] != 0]
stats_list = ['Name','Team','PA','AB','H','1B','2B','3B','HR','R','RBI','K','BB','HBP']
hits_list = ['single','home_run','triple','double']
print("BATTERS")
## Loop through the unique dates
for date in results_df['game_date'].unique():
    print(date)
    date_df = results_df.groupby('game_date').get_group(date)
    ## Loop through the home team
    for team in date_df['home_team'].unique():
        team_df = date_df.groupby('home_team').get_group(team)
        ## Loop through the inning_topbot
        for topbot in team_df['inning_topbot'].unique():
            topbot_df = team_df.groupby('inning_topbot').get_group(topbot)
            ## Loop through the players
            for player in topbot_df['batter'].unique():
                player_df = topbot_df.groupby('batter').get_group(player)
                ## Loop through game_pk (in event of double header)
                game_count = 1
                for game in player_df.game_pk.unique():
                    game_df = player_df[player_df['game_pk'] == game]
                    temp_player_df = pd.DataFrame(stats_list).T
                    temp_player_df.columns = stats_list
                    temp_player_df['Name'] = game_df['batter'].iloc[0]
                    if topbot == 'Bot':
                        temp_player_df['Team'] = game_df['home_team'].iloc[0]
                        temp_player_df['Opponent'] = game_df['away_team'].iloc[0]
                    else:
                        temp_player_df['Team'] = game_df['away_team'].iloc[0]
                        temp_player_df['Opponent'] = game_df['home_team'].iloc[0]
                    temp_player_df['K'] = len(game_df[game_df['events'] == 'strikeout']) + len(game_df[game_df['events'] == 'strikeout_double_play'])
                    temp_player_df['BB'] = len(game_df[game_df['events'] == 'walk'])
                    temp_player_df['PA'] = len(game_df)
                    temp_player_df['1B'] = len(game_df[game_df['events'] == 'single'])
                    temp_player_df['HR'] = len(game_df[game_df['events'] == 'home_run'])
                    temp_player_df['2B'] = len(game_df[game_df['events'] == 'double'])
                    temp_player_df['3B'] = len(game_df[game_df['events'] == 'triple'])
                    temp_player_df['H'] = temp_player_df['1B'] + temp_player_df['2B'] + temp_player_df['3B'] + temp_player_df['HR']
                    ## Initialize, since there will always be a R and RBI
                    temp_player_df['R'] = temp_player_df['HR']
                    temp_player_df['RBI'] = temp_player_df['HR']
                    ## Remove the walks from the PA to get a rough estimate of the AB's
                    temp_player_df['HBP'] = len(game_df[game_df['events'] == 'hit_by_pitch'])
                    temp_player_df['AB'] = temp_player_df['PA'] - temp_player_df['BB'] - temp_player_df['HBP']
                    temp_player_df['Game_Count'] = game_count
                    temp_player_df['Date'] = game_df['game_date'].iloc[0]
                    ## AB = len(game_df[game_df['events'] != BB, sac_bunt, sac_fly, sac_fly_double_play, catcher_interf
                    total_bat_df = pd.concat([total_bat_df,temp_player_df])
                    game_count += 1
print()
print()
```

```python
results_df = data.fillna(0)
total_pitch_df = pd.DataFrame()
stats_list = ['Name','Team','IP','H','R','K','BB','HBP']
hits_list = ['single','home_run','triple','double']
two_outs = ['grounded_into_double_play','double_play','strikeout_double_play','sac_fly_double_play']
one_out = ['field_out','strikeout','caught_stealing_2b','force_out','fielders_choice','sac_fly','fielders_choice_out','sac_bunt','other_out','pickoff_3b','caught_stealing_3b','pickoff_caught_stealing_2b','caught_stealing_home','pickoff_1b','pickoff_caught_stealing_home','pickoff_caught_stealing_3b','pickoff_2b']
## Loop through the unique dates
print("PITCHERS")
for date in results_df['game_date'].unique():
    print(date)
    date_df = results_df.groupby('game_date').get_group(date)
    ## Loop through the home team
    for team in date_df['home_team'].unique():
        team_df = date_df.groupby('home_team').get_group(team)
        ## Loop through the inning_topbot
        for topbot in team_df['inning_topbot'].unique():
            topbot_df = team_df.groupby('inning_topbot').get_group(topbot)
            ## Loop through the players
            game_count = 1
            for game in topbot_df.game_pk.unique():
                game_df = topbot_df[topbot_df['game_pk'] == game]
                player_num = len(game_df.groupby('pitcher').mean())
                for player in game_df['pitcher'].unique():
                    player_df = game_df.groupby('pitcher').get_group(player)
                    temp_player_df = pd.DataFrame(stats_list).T
                    temp_player_df.columns = stats_list
                    temp_player_df['Name'] = player_df['pitcher'].iloc[0]
                    if topbot == 'Top':
                        temp_player_df['Team'] = player_df['home_team'].iloc[0]
                        temp_player_df['Opponent'] = player_df['away_team'].iloc[0]
                        temp_player_df['R'] = player_df.iloc[0]['post_away_score'] - player_df.iloc[-1]['post_away_score']
                    else:
                        temp_player_df['Team'] = player_df['away_team'].iloc[0]
                        temp_player_df['Opponent'] = player_df['home_team'].iloc[0]
                        temp_player_df['R'] = player_df.iloc[0]['post_home_score'] - player_df.iloc[-1]['post_home_score']
                    temp_player_df['Game_Count'] = game_count
                    temp_player_df['K'] = len(player_df[player_df['events'] == 'strikeout']) + len(player_df[player_df['events'] == 'strikeout_double_play'])
                    temp_player_df['BB'] = len(player_df[player_df['events'] == 'walk'])
                    temp_player_df['IP'] = 0
                    new_temp_player_df = player_df[player_df['events'] !=0]
                    if len(new_temp_player_df) == 1:
                        outs = 1
                    else:
                        outs = 0
                        for x in range(len(new_temp_player_df['outs_when_up']) - 2, -1, -1):
                            if x == 0:
                                for out in one_out:
                                    if new_temp_player_df['events'].iloc[x] == out:
                                        outs += 1
                                        continue
                                for out in two_outs:
                                    if new_temp_player_df['events'].iloc[x] == out:
                                        outs += 2
                                        continue
                            if new_temp_player_df['outs_when_up'].iloc[x] == 0 and new_temp_player_df['outs_when_up'].iloc[x + 1] == 2:
                                outs += 1
                                continue
                            elif new_temp_player_df['outs_when_up'].iloc[x] == 0 and new_temp_player_df['outs_when_up'].iloc[x + 1] == 1:
                                outs += 2
                                continue
                            outs += new_temp_player_df['outs_when_up'].iloc[x] - new_temp_player_df['outs_when_up'].iloc[x + 1]
                    temp_player_df['Outs'] = outs
                    temp_player_df['IP'] = round(temp_player_df['Outs']/3,2)
                    if str(temp_player_df.iloc[0]['IP'])[-2] == '6':
                        temp_player_df['IP'] = int(temp_player_df['IP']) + 0.2
                    elif str(temp_player_df.iloc[0]['IP'])[-2] == '3':
                        temp_player_df['IP'] = int(temp_player_df['IP']) + 0.1
                    else:
                        temp_player_df['IP'] = int(temp_player_df['IP'])
                    temp_player_df['H'] = len(player_df[player_df['events'] == 'single']) + len(player_df[player_df['events'] == 'double']) + len(player_df[player_df['events'] == 'triple']) + len(player_df[player_df['events'] == 'home_run'])
                    temp_player_df['Pitches'] = len(player_df)
                    temp_player_df['Date'] = player_df['game_date'].iloc[0]
                    temp_player_df['HBP'] = len(player_df[player_df['events'] == 'hit_by_pitch'])
                    if player_num == 1:
                        temp_player_df['GS'] = 1
                    else:
                        temp_player_df['GS'] = 0
                        player_num -= 1
                    total_pitch_df = pd.concat([temp_player_df,total_pitch_df])
                game_count += 1
total_pitch_df.reset_index(inplace= True)
total_pitch_df.drop(columns = ['index'],inplace = True)
```
